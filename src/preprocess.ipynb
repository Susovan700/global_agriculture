```python
import pandas as pd

# Load dataset from raw folder
df = pd.read_csv("../raw/climate_change_impact_on_agriculture_2024.csv")

# Show first rows
df.head()
```

```python
# Check dataset details
df.info()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.fillna(df.mean(numeric_only=True))   # numeric columns
df = df.fillna("Unknown")                    # text columns

# Show few rows after cleaning
df.head()
```

```python
# Check missing values in each column
df.isnull().sum()

# Example: Fill missing numeric values with mean
df = df.fillna(df.mean(numeric_only=True))

# Example: Drop rows where important categorical columns are missing
# (replace 'column_name' with actual)
# df = df.dropna(subset=['column_name'])

df.head()
```

```python
# Clean column names (lowercase, replace spaces with _)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.head()
```

```python
df.to_csv("../processed/climate_change_impact_on_agriculture_2024_cleaned.csv", index=False)
print("✅ Preprocessed data saved successfully!")
```
